## Question 7 (3 points)

**This is the main challenge: Extract all dates from the medical notes and sort them chronologically.**

**Rules:**
- Assume all dates in `xx/xx/xx` format are `mm/dd/yy`
- Assume all 2-digit years are from the 1900s (e.g., `1/5/89` is January 5th, 1989)
- If the day is missing (e.g., `9/2009`), assume it is the 1st day of the month
- If the month is missing (e.g., `2010`), assume it is January 1st

*This function should return a pandas Series of length 500, where the values are the original indices sorted by date in ascending chronological order.*

**Example:**
```python
# If original series was:
#    0    1999
#    1    2010
#    2    1978
# Your function should return:
#    0    2    (1978 is earliest)
#    1    0    (1999 is second)
#    2    1    (2010 is latest)
```

````python
def question_seven():
    """
    Extract dates from all medical notes and return indices sorted chronologically.

    Returns:
        pd.Series: Series of length 500 with original indices sorted by date
    """
    dates = []

    for idx, note in enumerate(df):
        # Find the earliest date in this note
        extracted_date = extract_date_from_note(note)
        dates.append((idx, extracted_date))

    # Sort by date (second element of tuple)
    sorted_dates = sorted(dates, key=lambda x: x[1])

    # Extract just the indices
    sorted_indices = [idx for idx, date in sorted_dates]

    return pd.Series(sorted_indices)


def extract_date_from_note(text):
    """
    Extract the earliest date from a medical note text.
    Returns a datetime object.

    # Hint:
    # 1. Create regex patterns to match different date formats
    # 2. Extract dates from each note
    # 3. Parse dates into datetime objects
    # 4. Sort by date and return the indices
    """
    dates_found = []

    # Pattern 1: mm/dd/yy format (e.g., 1/5/89, 01/05/89)
    pattern1 = r'\b(\d{1,2})/(\d{1,2})/(\d{2})\b'
    for match in re.finditer(pattern1, text):
        month, day, year = match.groups()
        month, day, year = int(month), int(day), int(year)
        # Assume 2-digit years are from 1900s
        year = 1900 + year
        try:
            dates_found.append(datetime(year, month, day))
        except ValueError:
            # Invalid date, skip
            pass

    # Pattern 2: m/yyyy format (e.g., 9/2009, 01/2010) - assume day is 1st
    pattern2 = r'\b(\d{1,2})/(\d{4})\b'
    for match in re.finditer(pattern2, text):
        month, year = match.groups()
        month, year = int(month), int(year)
        try:
            dates_found.append(datetime(year, month, 1))
        except ValueError:
            pass

    # Pattern 3: yyyy format (e.g., 2010, 1999) - assume January 1st
    pattern3 = r'\b(\d{4})\b'
    for match in re.finditer(pattern3, text):
        year = int(match.group(1))
        # Only consider 4-digit numbers that look like years (1800-2100)
        if 1800 <= year <= 2100:
            try:
                dates_found.append(datetime(year, 1, 1))
            except ValueError:
                pass

    # Return the earliest date found, or a very old default if none found
    if dates_found:
        return min(dates_found)
    else:
        return datetime(1900, 1, 1)


# Test your function
q7_result = question_seven()
print(f"Result length: {len(q7_result)}")
print(f"First 10 indices: {list(q7_result.head(10))}")
print(f"Last 10 indices: {list(q7_result.tail(10))}")
````
